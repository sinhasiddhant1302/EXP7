```python
#Print i as long as i is less than 6:
i=1
while i<6:
  print(i)
  i+=1
```

```python
n=int(input("Enter a number:"))
i=1
while i<=n:
  print(i)
  i=i+1
```

```python
#Factorial
n=int(input("Enter a number:"))
fact=1

while n>1:
  fact=fact*n                   #also can be written as fact *= n
  n-= 1
print("FACTORIAL",fact)
```

```python
#Fibonacci Series using while loop

n=int(input("Enter number of terms:"))

a=0
b=1
i=1
while i<= n:
  print(a,end=" ")      #end makes the series horizontal
  c=a+b
  a=b
  b=c
  i += 1
```

```python
limit=int(input("Enter the limit:"))
a=0
b=1
while a <= limit:
  print(a, end= " ")
  a, b=b, a+b
```

```python
#Reverse a number
num = int(input("Enter a number: "))
rev = 0
while num > 0:
 digit = num % 10
 rev = rev * 10 + digit
 num //= 10                       #// removes last number

print("Reversed Number = ",rev)
```

```python
#Check Palidrome number
num= int(input("Enter a number:"))
temp = num
rev = 0
while num > 0:
  rev= rev * 10 + num % 10
  num //= 10
if temp == rev:
    print("Palidrome")
else:
    print("Not Pallidrome")
```

```python
#Check Palidrome for String
s = (input("Enter a string:"))
i, j = 0, len(s) - 1
is_palidrome = True
while i < j:
  if s[i] != s[j]:
    is_palidrome = False
    break
  i += 1
  j -= 1
if is_palidrome:
    print("Palidrome")
else:
    print("Not Palidrome")
```

```python
#Palidrome using if loop only
st= input("Enter a string:")
if st==st[::-1]:
  print("Palidrome")
else:
  print("Not Palidrome")
```

```python
#Count number of digits in a given number using while loop
num = int(input("Enter a number:"))        #very easy by using len()
count = 0
while num > 0:
  count += 1
  num //= 10
print ("Number of digits= ",count)
```

```python
#Search an element in a list
nums = [10,20,30,40,50]
key = int(input("Enter an element to search:"))
i = 0
while i< len(nums):
  if nums[i] == key:
    print("Element found at index",i)
    break
  i += 1
else:
  print("Element not found")
```
